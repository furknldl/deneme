```python
import requests
from bs4 import BeautifulSoup

# KullanÄ±cÄ±dan web sayfasÄ±nÄ±n URL'sini al
url = input("LÃ¼tfen web sayfasÄ±nÄ±n URL'sini girin: ")

# Web sayfasÄ±nÄ± getir
response = requests.get(url)

# Ä°steÄŸin baÅŸarÄ±lÄ± olup olmadÄ±ÄŸÄ±nÄ± kontrol et
if response.status_code == 200:
    # HTML iÃ§eriÄŸini iÅŸle
    html_content = response.content
    soup = BeautifulSoup(html_content, 'html.parser')

    # TÃ¼m <a> etiketlerini bul
    links = soup.find_all('a')

    # Her bir link iÃ§in URL ve metni yazdÄ±r
    for link in links:
        url = link.get('href')
        text = link.get_text(strip=True)
        print("URL:", url)
        print("Metin:", text)
        print()

else:
    print("Web sayfasÄ±na eriÅŸilemiyor.")

```
